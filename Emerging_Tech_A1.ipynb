### Import necessary libraries

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.preprocessing import StandardScaler
```

### Load the model and the dataset

```python
# Load the model
reg = linear_model.LinearRegression()

# Note that X must be 2D
X = np.array([[22],[25],[27],[30],[33]])
y = np.array([120,150,170,210,240])

# Here we do not need the test data
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=None, shuffle=False)

X_train = X
y_train = y

# Standardize the features
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_for_pred_scaled = scaler.transform([[28]]) # The input value for prediction given in question
```

### Fit the model to the dataset

```python
reg.fit(X_train_scaled,y_train)
```

### Predict the value

```python
y_pred = reg.predict(X_for_pred_scaled)
print(y_pred)
```

```python
print(f"slope = {reg.coef_[0]}\nintercept = {reg.intercept_}")
```
