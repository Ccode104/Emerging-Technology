### Import necessary libraries

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import numpy as np
```

### Load the model and the dataset

```python
reg = linear_model.LinearRegression()

X = np.array([850,900,1000,1200,1500])
y = np.array([50,55,65,80,95])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=2, shuffle=False)
```

### Fit the model to the dataset

```python
# We must reshape the arrays to make it 2D even if it has only 1 feature
reg.fit(np.array(X_train).reshape(-1, 1),np.array(y_train).reshape(-1, 1))
```

```python
from sklearn.metrics import mean_squared_error ,r2_score

y_pred = reg.predict(np.array([1100]).reshape(-1, 1))
print(y_pred)
```
